**Extract text**

```python
import pdfplumber

def extract_text_from_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:  # avoid None
                text += page_text + "\n"
    return text
```

**Basic cleaning (important)**

```python
import re

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)   # remove extra spaces
    text = re.sub(r'\n+', '\n', text)  # normalize newlines
    return text.strip()

cleaned_text = clean_text(raw_text)

print("\n--- CLEANED TEXT ---\n")
print(cleaned_text[:2000])
```

# **ROLE CONFIG**

```python

roles = {
    "frontend": {
        "must_have": ["html", "css", "javascript"],
        "good_to_have": ["react", "tailwind", "redux"]
    },
    "backend": {
        "must_have": ["node", "api", "database"],
        "good_to_have": ["express", "mongodb", "authentication"]
    },
    "data": {
        "must_have": ["python", "sql"],
        "good_to_have": ["machine learning", "pandas", "numpy"]
    }
}
```

**TEXT NORMALIZATION**

```python

def normalize_text(text):
    return text.lower()
```

**SKILL ALIAS MAPPING**

```python

skill_aliases = {
    "javascript": [],
    "react": ["react.js"],
    "node": ["node.js"],
    "machine learning": ["ml"]
}
```

**CHECK SKILL IN TEXT**

```python

def skill_in_text(skill, text):
    if skill in text:
        return True

    # check aliases
    if skill in skill_aliases:
        for alias in skill_aliases[skill]:
            if alias in text:
                return True

    return False
```

**MAIN SCORING FUNCTION**

```python

def analyze_resume(resume_text, role):
    text = normalize_text(resume_text)

    config = roles.get(role)
    if not config:
        return {"error": "Invalid role"}

    matched = []
    missing = []

    score = 0

    # scoring weights
    MUST_WEIGHT = 3
    GOOD_WEIGHT = 1

    # check must-have skills
    for skill in config["must_have"]:
        if skill_in_text(skill, text):
            matched.append(skill)
            score += MUST_WEIGHT
        else:
            missing.append(skill)

    # check good-to-have skills
    for skill in config["good_to_have"]:
        if skill_in_text(skill, text):
            matched.append(skill)
            score += GOOD_WEIGHT

    return {
        "role": role,
        "score": score,
        "matched_skills": matched,
        "missing_skills": missing
    }
```

```python
print(analyze_resume(cleaned_text, "frontend"))
print(analyze_resume(cleaned_text, "backend"))
print(analyze_resume(cleaned_text, "data"))
```

# **Resume vs Job Description matching**

**(Reuse) Skill aliases**

```python
skill_aliases = {
    "javascript": [],
    "react": ["react.js"],
    "node": ["node.js"],
    "machine learning": ["ml"]
}
```

**Skill check function (same as before)**

```python
def skill_in_text(skill, text):
    if skill in text:
        return True

    if skill in skill_aliases:
        for alias in skill_aliases[skill]:
            if alias in text:
                return True

    return False
```

**Master skill list (IMPORTANT)**

```python
ALL_SKILLS = [
    "html", "css", "javascript", "react", "redux",
    "node", "express", "mongodb", "api", "database",
    "python", "sql", "machine learning", "pandas", "numpy"
]
```

**Extract skills from JD**

```python
def extract_skills_from_jd(jd_text):
    jd_text = jd_text.lower()

    required_skills = []

    for skill in ALL_SKILLS:
        if skill_in_text(skill, jd_text):
            required_skills.append(skill)

    return required_skills
```

**Compare Resume vs JD**

```python
def analyze_resume_vs_jd(resume_text, jd_text):
    resume_text = resume_text.lower()

    required_skills = extract_skills_from_jd(jd_text)

    matched = []
    missing = []

    for skill in required_skills:
        if skill_in_text(skill, resume_text):
            matched.append(skill)
        else:
            missing.append(skill)

    # avoid division by zero
    if len(required_skills) == 0:
        match_percentage = 0
    else:
        match_percentage = int((len(matched) / len(required_skills)) * 100)

    return {
        "required_skills": required_skills,
        "matched_skills": matched,
        "missing_skills": missing,
        "match_percentage": match_percentage
    }
```

**Test it**

```python
jd_text = """
Looking for a frontend developer with React, JavaScript, CSS, Redux experience.
"""

result = analyze_resume_vs_jd(cleaned_text, jd_text)

print(result)
```

**Combine everything**

```python
import re

def clean_feedback(feedback_text):
    lines = feedback_text.split("\n")

    cleaned = []

    for line in lines:
        line = line.strip()

        if line.startswith("*") or line.startswith("-"):
            # remove bullet symbols
            line = re.sub(r"^[\*\-\•]\s*", "", line)

            # remove markdown bold (**text**)
            line = re.sub(r"\*\*(.*?)\*\*", r"\1", line)

            cleaned.append(line)

    return cleaned
```

```python
def build_final_output(analysis_result, feedback_text):

    feedback_list = clean_feedback(feedback_text)

    return {
        "match_percentage": analysis_result["match_percentage"],
        "matched_skills": analysis_result["matched_skills"],
        "missing_skills": analysis_result["missing_skills"],
        "feedback": feedback_list
    }
```
